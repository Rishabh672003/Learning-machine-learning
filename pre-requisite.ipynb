```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x = np.linspace(-6,6,1000)
```

```python
fx = norm.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x,fx);
```

```python
fx = norm.cdf(x,loc=0,scale = 1);
plt.plot(x,fx);
```
